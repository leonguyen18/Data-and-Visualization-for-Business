```python
# Check the distribution of alert levels
print("Alert levels distribution:")
print(df['alert'].value_counts(dropna=False))

# Check the relationship between alert and magnitude
print("\nMean magnitude for each alert level:")
print(df.groupby('alert')['magnitude'].mean())

# Check the distribution of magnitude for each alert level
print("\nMagnitude distribution by alert level:")
print(df.groupby('alert')['magnitude'].describe())
```

The relationship between magnitude and alert:
- green: mean = 5.38, range from 3.5 to 7.1
- yellow: mean = 6.20, range from 5.3 to 7.6
- orange: mean = 6.30, all values are 6.3
- red: mean = 6.78, range from 5.4 to 7.5

```python
# Determine earthquake alert level based on magnitude
def determine_alert_level(row):
    """
    Determine earthquake alert level based on magnitude.
    Only determine new alert level if current alert is missing.
    """
    # Check if alert is missing
    if pd.isna(row['alert']):
        magnitude = row['magnitude']
        
        # Low impact earthquakes (< 5.0)
        if magnitude < 5.0:
            return 'green'
        
        # Light to moderate earthquakes (5.0 - 5.9)
        if magnitude >= 5.0 and magnitude < 6.0:
            if magnitude < 5.5:
                # Light earthquakes (5.0 - 5.4)
                return 'green'
            else:
                # Moderate earthquakes (5.5 - 5.9)
                return 'yellow'
        
        # Strong earthquakes (6.0 - 6.4)
        if magnitude >= 6.0 and magnitude < 6.5:
            return 'yellow'
        
        # Major earthquakes (6.5 - 6.9)
        if magnitude >= 6.5 and magnitude < 7.0:
            return 'orange'
        
        # Great earthquakes (7.0+)
        if magnitude >= 7.0:
            return 'red'
    else:
        return row['alert']

# Apply function to entire dataframe
df['alert'] = df.apply(determine_alert_level, axis=1)
```

```python
# Verify the distribution after filling
print("New alert level distribution:")
print(df['alert'].value_counts(dropna=False))
```

```python
# Check min and max dates in the dataset
print("Earliest date:", earthquake['Date'].min())
print("Latest date:", earthquake['Date'].max())
```

```python
# Count earthquakes by month
monthly_counts = earthquake['Month'].value_counts().sort_index()
print("\nNumber of earthquakes by month:")
print(monthly_counts)
```

```python
# Count earthquakes in United States
us_earthquakes = earthquake[earthquake['Country'] == 'United States']
print("\nNumber of earthquakes in United States:", len(us_earthquakes))
```

```python
# Calculate average magnitude by country and sort in descending order
avg_magnitude_by_country = earthquake.groupby('Country')['Magnitude'].mean().sort_values(ascending=False)
print("\nAverage earthquake magnitude by country:")
print(avg_magnitude_by_country)
```

```python
# Calculate total felt earthquakes by country and sort in descending order
felt_by_country = earthquake.groupby('Country')['Felt'].sum().sort_values(ascending=False)
print("\nNumber of people who felt earthquakes by country:")
print(felt_by_country)
```

```python
# Count earthquakes and tsunamis by continent
tsunami_by_continent = earthquake.groupby('Continent')[['Tsunami']].sum()
earthquake_counts = earthquake['Continent'].value_counts()

# Combine into a single DataFrame
continent_summary = pd.DataFrame({
    'Total_Earthquakes': earthquake_counts,
    'Tsunamis': tsunami_by_continent['Tsunami']
}).sort_values('Total_Earthquakes', ascending=False)

print("\nEarthquakes and tsunamis by continent:")
print(continent_summary)
```
